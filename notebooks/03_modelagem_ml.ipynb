```python
import pandas as pd

caminho = "/content/drive/MyDrive/Projeto final(ESI)/dados/metricas_git_tratadas.csv"
df = pd.read_csv(caminho)

df.head()
```

```python
X = df[["adicoes", "remocoes", "num_commits"]]
y = df["mudou"]
```

```python
#treino
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

```python
#modelo
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=5,
    random_state=42
)
dt.fit(X_treino, y_treino)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_treino, y_treino)
```

```python
from sklearn.metrics import classification_report

print("Árvore de Decisão")
print(classification_report(y_teste, dt.predict(X_teste)))

print("Regressão Logística")
print(classification_report(y_teste, lr.predict(X_teste)))
```

```python
modelo_campeao = dt
```
