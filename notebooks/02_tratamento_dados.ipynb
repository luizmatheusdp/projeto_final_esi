```python
import pandas as pd

caminho = "/content/drive/MyDrive/Projeto final(ESI)/dados/metricas_git_brutas.csv"
df = pd.read_csv(caminho)

df.head()
```

```python
# Remover os sem nome de arquivo
df = df.dropna(subset=["arquivo"])

# Remover arquivos muito gen√©ricos
df = df[df["arquivo"] != ""]

df.info()
```

```python
df_agg = df.groupby("arquivo").agg({
    "adicoes": "sum",
    "remocoes": "sum",
    "hash_commit": "nunique"
}).reset_index()

df_agg.rename(columns={
    "hash_commit": "num_commits"
}, inplace=True)

df_agg.head()
```

```python
df_agg["mudou"] = (df_agg["num_commits"] > 1).astype(int) #Variavel alvo (label)
```

```python
saida = "/content/drive/MyDrive/Projeto final(ESI)/dados/metricas_git_tratadas.csv"
df_agg.to_csv(saida, index=False)

print("Dados tratados salvos com sucesso!")
```
