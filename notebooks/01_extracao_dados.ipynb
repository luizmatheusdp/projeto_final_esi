```python
from pydriller import Repository
import pandas as pd

repo_path = "https://github.com/psf/requests.git"

dados = []

for commit in Repository(repo_path).traverse_commits():
    for mod in commit.modified_files:
        dados.append({
            "hash_commit": commit.hash,
            "autor": commit.author.name,
            "data_commit": commit.committer_date,
            "arquivo": mod.filename,
            "adicoes": mod.added_lines,
            "remocoes": mod.deleted_lines,
            "loc": mod.nloc,
            "complexidade": mod.complexity,
            "tipo_modificacao": mod.change_type.name
        })

df = pd.DataFrame(dados)
df.head()
```

```python
df.to_csv("metricas_git_brutas.csv", index=False)
print("CSV bruto salvo com sucesso")
```
